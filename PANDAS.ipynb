```python
df=pd.DataFrame({"name":['abhinav','andrie','abhinav'],
      "marks":[np.nan,np.nan,np.nan],
      "country":['india','romania',pd.NaT]})
```

```python
df.head()
```

```python
df.dropna()
#for dropping the nan value
```

```python
df.dropna(how='all',axis=1)
#this is basiclaay for remove the columns if all the variable are null
```

```python
df.drop_duplicates(subset=['name'],keep=False)
#to delete the duplicate elemnt keep for if u eant to keep single duplicate elemnt or not
```
